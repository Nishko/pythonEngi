Calculated before child puts in $120 each month.

```python
initial_balance = 800
monthly_contribution = 120
target_balance = 40000

balance = initial_balance
interest_rate = 0.003
months = 0

balances_over_time = []

while balance < target_balance:
    interest = balance * interest_rate

    balance += interest + monthly_contribution

    months += 1

    balances_over_time.append(balance)

print(
    f"It will take {months} months to save ${target_balance} for the child's education."
)
print("Monthly Balances:")
for i, monthly_balance in enumerate(balances_over_time, start=1):
    print(f"Month {i}: ${monthly_balance:.2f}")
```

Answer for interest calculated for beginning of the month

```python
initial_balance = 800
monthly_contribution = 120
target_balance = 40000

balance = initial_balance
months = 0

balances_over_time = []

while balance < target_balance:
    if balance < 10000:
        interest_rate = 0.003  # 0.3% per month
    elif balance < 20000:
        interest_rate = 0.004  # 0.4% per month
    elif balance < 30000:
        interest_rate = 0.005  # 0.5% per month
    else:
        interest_rate = 0.006  # 0.6% per month

    interest = balance * interest_rate

    balance += monthly_contribution + interest

    months += 1

    balances_over_time.append(balance)

print(
    f"It will take {months} months to save ${target_balance} for the child's education."
)
print("Monthly Balances:")
for i, monthly_balance in enumerate(balances_over_time, start=1):
    print(f"Month {i}: ${monthly_balance:.2f}")
```
