```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def g(x):
    # Calculate g(x) = (e^(2x) - e^(-2x)) / 2
    # Parameters:
    # - x : scalar or array-like (Input values)
    # Returns:
    # - Scalar or array-like of g(x) values
    return (np.exp(2 * x) - np.exp(-2 * x)) / 2


def h(x):
    # Calculate h(x) = (e^(2x) - e^(-2x)) / (e^(2x) + e^(-2x))
    # Parameters:
    # - x : scalar or array-like (Input values)
    # Returns:
    # - Scalar or array-like of h(x) values
    return (np.exp(2 * x) - np.exp(-2 * x)) / (np.exp(2 * x) + np.exp(-2 * x))


def plot_functions():
    # Create values for x from -5 to 5
    x_values = np.linspace(-5, 5, 400)

    # Compute g(x) and h(x) for these values
    g_values = g(x_values)
    h_values = h(x_values)

    # Plot the functions
    plt.figure(figsize=(10, 6))
    plt.plot(x_values, g_values, label="g(x)")
    plt.plot(x_values, h_values, label="h(x)")
    plt.title("Plot of g(x) and h(x)")
    plt.xlabel("x")
    plt.ylabel("Value")
    plt.legend()
    plt.grid(True)
    plt.show()


# Call the plotting function
plot_functions()
```
