```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read flow rate values from file
flow_rate = np.loadtxt("wk9_flowrate.dat")

# Initial volume of the reservoir in m3
initial_volume = 1000.0

# Initialize list to store the volume at each hour
volume_list = [initial_volume]

# Initialize the volume
current_volume = initial_volume

# Calculate the volume at each hour
for i in range(len(flow_rate)):
    # Calculate the volume of water coming in
    Qin = flow_rate[i]

    # Calculate the volume of water going out
    Qout = 10 * np.random.rand()

    # Update the current volume
    current_volume += Qin - Qout

    # Append to list
    volume_list.append(current_volume)

# Calculate the final volume
final_volume = volume_list[-1]

# Generate time (in hours) for plotting
time_hours = np.arange(0, len(flow_rate) + 1)

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(time_hours, volume_list, marker="o", linestyle="-")
plt.title("Volume of Water in Reservoir Throughout the Day")
plt.xlabel("Time (hours)")
plt.ylabel("Volume (m^3)")
plt.grid(True)
plt.show()

# Display the final volume
print(f"The volume of the reservoir at the end of the day is {final_volume:.2f} m^3.")
```
